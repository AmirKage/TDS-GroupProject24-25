```python
import json

# Define labels for each question
q1_labels = ["Salesforce", "HubSpot", "Pipedrive", "Microsoft Dynamics", "Close.io", "SAP Sales Cloud"]
q2_labels = [
    "Lead management", "Task automation", "Email tracking", "Contact management",
    "Analytics dashboard", "Pipeline visualization", "Meeting scheduling", "Custom reporting"]
q3_labels = [
    "Schedule Coordination", "Team Efficiency Balance", "Schedule Changes", "Budget Adherence",
    "Promotional Material Management", "Compliance Adherence", "ROI Measurement", "Language Barriers"]
q4_labels = [
    "User Interface & Usability", "AI & Data Insights", "Third-party Integration",
    "Mobile & Offline Functionality", "Customization & Personalization", "Collaboration & Scalability"]
q5_labels = [
    "Data Audits & Cleansing", "Automation & AI Integration", "Data Validation & Accuracy",
    "Data Consistency & Syncing", "Training & Best Practices", "Collaboration & Monitoring"]
q6_labels = [
    "AI & Predictive Tools", "Team Collaboration & Communication", "Automation & Data Enrichment",
    "Customer Insights & Feedback", "Customization & Flexibility", "Cross-Platform Integration"]

# Map questions to their labels
labels = {
    "Q1": q1_labels,
    "Q2": q2_labels,
    "Q3": q3_labels,
    "Q4": q4_labels,
    "Q5": q5_labels,
    "Q6": q6_labels
}

# Load the tokenized dataset
with open("structured_tokenized_data.json", "r") as file:
    dataset = json.load(file)

# Function to map responses to numerical labels
def map_response_to_label(answer, label_list):
    for idx, label in enumerate(label_list):
        if label.lower() in answer.lower():  # Check if label exists in the response
            return idx
    return -1  # Return -1 if no label matches

# Add labels to the dataset
for item in dataset:
    question = item["question"]
    answer = item["answer"]
    if question in labels:  # Ensure the question has a corresponding label list
        item["label"] = map_response_to_label(answer, labels[question])
    else:
        item["label"] = -1  # Default to -1 for unmatched questions

# Save the labeled dataset back to a JSON file
with open("labeled_dataset.json", "w") as file:
    json.dump(dataset, file, indent=4)

print("Labeled dataset has been saved to labeled_dataset.json")
```

```python
# Define function to map responses to labels
def map_responses_to_labels(responses, labels):
    numeric_labels = []
    for response in responses:
        matched = False
        for idx, label in enumerate(labels):
            if label.lower() in response.lower():
                numeric_labels.append(idx)
                matched = True
                break
        if not matched:
            numeric_labels.append(-1)  # Assign -1 if no label matches
    return numeric_labels

# Store all responses in a dictionary with labels and numeric labels
responses_data = {}
for question, data in zip(responses.keys(), labels.values()):
    question_responses = responses[question]
    question_labels = data
    numeric_labels = map_responses_to_labels(question_responses, question_labels)
    responses_data[question] = {
        "responses": question_responses,
        "labels": question_labels,
        "numeric_labels": numeric_labels
    }

# Print the final dataset structure
for question, data in responses_data.items():
    print(f"{question}:")
    print("Responses:", data["responses"])
    print("Labels:", data["labels"])
    print("Numeric Labels:", data["numeric_labels"])
```

```python
import torch
from torch.utils.data import Dataset

# Custom Dataset class
class CRMResponsesDataset(Dataset):
    def __init__(self, tokenized_data, labels=None):
        self.input_ids = tokenized_data["input_ids"]
        self.attention_mask = tokenized_data["attention_mask"]
        self.labels = labels if labels else [None] * len(self.input_ids)

        # Ensure labels and input_ids are of the same length
        if len(self.labels) != len(self.input_ids):
            if len(self.labels) < len(self.input_ids):
                self.labels.extend([None] * (len(self.input_ids) - len(self.labels)))
            else:
                self.labels = self.labels[:len(self.input_ids)]

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        item = {
            "input_ids": self.input_ids[idx],
            "attention_mask": self.attention_mask[idx],
        }
        if self.labels[idx] is not None:
            item["labels"] = torch.tensor(self.labels[idx])
        return item

# Function to map responses to labels
def map_response_to_label(response, labels):
    matched_labels = [i for i, label in enumerate(labels) if label.lower() in response.lower()]
    if len(matched_labels) == 1:
        return matched_labels[0]
    elif len(matched_labels) > 1:
        print(f"Warning: Multiple matches found for response: {response}. Using first match.")
        return matched_labels[0]
    else:
        return -1

# Create datasets for all questions
datasets = {}
for question, question_labels in labels.items():
    numeric_labels = [map_response_to_label(response, question_labels) for response in responses[question]]
    datasets[question] = CRMResponsesDataset(tokenized_data[question], labels=numeric_labels)

# Verify dataset lengths
for question, dataset in datasets.items():
    print(f"{question}: {len(dataset)} samples")
```

```python
import torch
from torch.utils.data import Dataset

class CRMResponsesDataset(Dataset):
    def __init__(self, tokenized_data, labels=None):
        self.input_ids = tokenized_data["input_ids"]
        self.attention_mask = tokenized_data["attention_mask"]

        # Ensure labels list has the same length as input data
        # If labels are provided, use them. Otherwise, create a list of None values with the correct length.
        self.labels = labels if labels else [None] * len(self.input_ids)

        # Adjust labels length to match input_ids length if necessary
        if len(self.labels) != len(self.input_ids):
            # Extend the labels list with None values if it's shorter
            if len(self.labels) < len(self.input_ids):
                self.labels.extend([None] * (len(self.input_ids) - len(self.labels)))
            # Truncate the labels list if it's longer
            else:
                self.labels = self.labels[:len(self.input_ids)]

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        item = {
            "input_ids": self.input_ids[idx],
            "attention_mask": self.attention_mask[idx],
        }
        if self.labels[idx] is not None:
            item["labels"] = torch.tensor(self.labels[idx])
        return item

# Example dataset for Question 1
q1_dataset = CRMResponsesDataset(
    tokenized_data["Q1"],
    labels=[next((i for i, label in enumerate(q1_labels) if label.lower() in response.lower()), -1)
            for response in responses["Q1"]]
)

q2_dataset = CRMResponsesDataset(
    tokenized_data["Q2"],
    labels=[next((i for i, label in enumerate(q2_labels) if label.lower() in response.lower()), -1)
            for response in responses["Q2"]]
)
```
