```python
import json
import random

# Define the questions
questions = [
    "Describe an event that had a profound impact on your life or personal growth and how it changed your perspective on the world.",
    "What skill or trait has been most important to your success in the workplace? Please explain why.",
    "Which technology do you think will have the greatest impact on society in the next 10 years, and why?"
]

# Define response pools for each question, expanded to ensure at least 100 unique responses
response_pools = {
    questions[0]: [
        "When I traveled abroad alone for the first time, I learned how to adapt to new environments and gained confidence in myself.",
        "Losing a loved one made me realize how important it is to cherish the moments we have with our family.",
        "Starting my own business taught me resilience and how to handle failure.",
        "Participating in a volunteer program opened my eyes to the struggles others face and inspired me to give back to my community.",
        "Overcoming a serious illness changed my outlook on life and made me value my health and relationships more.",
        "Completing a marathon showed me the power of persistence and self-discipline.",
        "Moving to a new city where I didn’t know anyone helped me develop independence and social skills.",
        "Volunteering at an animal shelter taught me empathy and the importance of helping those in need.",
        "Joining a debate club improved my critical thinking and public speaking skills.",
        "Learning to play a musical instrument taught me patience and the joy of mastering a new skill."
    ] * 10,  # Expand responses by repeating
    questions[1]: [
        "Being a good communicator has helped me collaborate effectively with my team and resolve conflicts.",
        "Time management has been crucial for balancing multiple projects and meeting deadlines consistently.",
        "Adaptability has allowed me to navigate changes in the workplace and seize new opportunities.",
        "Having a growth mindset has encouraged me to continuously learn and improve my skills.",
        "Empathy has helped me understand my colleagues' perspectives and build stronger working relationships.",
        "Problem-solving skills have enabled me to address challenges creatively and find effective solutions.",
        "Networking has allowed me to build valuable professional relationships and opportunities.",
        "Leadership qualities have helped me inspire and guide my team toward achieving shared goals.",
        "Attention to detail has ensured the accuracy and quality of my work.",
        "Perseverance has helped me overcome obstacles and stay focused on long-term objectives."
    ] * 10,
    questions[2]: [
        "Artificial intelligence will revolutionize industries by automating tasks and improving decision-making.",
        "Renewable energy technologies, like solar and wind, will be crucial for combating climate change and ensuring sustainable development.",
        "Blockchain technology has the potential to enhance transparency and security in various sectors, including finance and supply chains.",
        "Advancements in biotechnology will lead to breakthroughs in healthcare, such as personalized medicine and gene editing.",
        "The Internet of Things will transform how we interact with our environment by connecting everyday devices and enabling smarter living.",
        "Quantum computing will enable solving complex problems that are currently beyond our reach.",
        "Augmented reality will enhance education and training by providing immersive learning experiences.",
        "5G technology will revolutionize communication and enable faster and more reliable connectivity.",
        "Robotics will take over repetitive tasks and assist in hazardous environments, improving efficiency and safety.",
        "Autonomous vehicles will change the way we travel and reduce traffic accidents caused by human error."
    ] * 10
}

# Generate unique responses for each question
def generate_unique_responses(question, num_responses=100):
    response_pool = response_pools[question]
    if len(response_pool) < num_responses:
        raise ValueError(f"Not enough unique responses available for the question: {question}")
    responses = random.sample(response_pool, num_responses)  # Ensure no duplicates
    return responses

# Compile data into a JSON-compatible structure
data = []
for question in questions:
    question_data = {
        "question": question,
        "responses": generate_unique_responses(question)
    }
    data.append(question_data)

# Save the dataset to a JSON file
output_file = "unique_responses_dataset.json"
with open(output_file, "w") as file:
    json.dump(data, file, indent=4)

# Print a success message and show a sample of the data
print(f"Dataset with {len(questions) * 100} unique responses has been saved to {output_file}.")
print(json.dumps(data[:1], indent=4))  # Print a sample
```
